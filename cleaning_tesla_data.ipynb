```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt



#loading dataset 

data =pd.read_csv('tesla_q1_2024.csv')

print("data Columns:")
print(data.columns.tolist())
```

```python
# Iterate through each column in the dataset
for column_name, series in data.items():
    # Calculate missing values for each column
    missing_values = series.isnull().sum()
    
    # Check if the column has any missing values
    if missing_values > 0:
        print(f"\nMissing values in '{column_name}': {missing_values}")

     
```

```python

#drop the null values 
data=data.dropna()
```

```python
# Drop rows with any null values
data_cleaned = data.dropna()

# Check the cleaned dataset for any remaining missing values
missing_values_after_drop = data_cleaned.isnull().sum()
non_zero_missing_values_after_drop = missing_values_after_drop[missing_values_after_drop > 0]

# Print if there are still any missing values
if not non_zero_missing_values_after_drop.empty:
    print("There are still missing values in the dataset:")
    print(non_zero_missing_values_after_drop)
else:
    print("No missing values remain in the dataset after dropping rows.")
    
# Optionally, you can display the first few rows of the cleaned data
print(data_cleaned.head())
```

```python
#save csv file 
data.to_csv("cleanned_tesla_q1_2024.csv", index=False)
```
